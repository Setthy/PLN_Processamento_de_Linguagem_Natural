# NLP - Mineração de textos.

```python
# Importando as bibliotecas:

import matplotlib.pyplot as plt
import nltk

# nltk.download() (pode ser executado apos instalar o nltk - pip install nltk)

nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud

# conda install -c conda-forge wordcloud (executar no anaconda prompt)

import string
```

```python
# Criação de corpus lendo textos do disco:

corpus = PlaintextCorpusReader('Arquivos', '.*', encoding = 'ISO-8859-1')
```

```python
# Leitura dos arquivos do disco, percorrer os registros e mostrar o nome dos primeiros 100 arquivos:

arquivos = corpus.fileids()

#primeiro arquivo:

arquivos[0]
```

```python
# zero a 10:

arquivos[0:10]
```

```python
# imprime todos os nomes:

for a in arquivos:
    print(a)
```

```python
# Acesso ao texto do primeiro arquivo:

texto = corpus.raw('1.txt')
texto
```

```python
# Acesso a todas as palavras de todos os arquivos do corpus:

todo_texto = corpus.raw()
#todo_texto
```

```python
# Obtenção de todos as palavras do corpus e visualização da quantidade:

palavras = corpus.words()

# Acessando pelo indice:

palavras[170]
```

```python
# Quantidade:

len(palavras)
```

```python
# Usando o NLTK, obtemos as stopword em ingles:

stops = stopwords.words('english')

#stops = stopwords.words('portuguese')

stops
```

```python
# Definição das cores que serão utilizadas na nuvem de pelavras:

mapa_cores = ListedColormap(['orange', 'green', 'red', 'magenta'])

# Criação da nuvem de palavras, com no maximo 100 palavras e utilizando as stop words:

nuvem = WordCloud(background_color = 'white', colormap = mapa_cores, stopwords = stops, max_words = 100)

# Criação e visualização da nuvem de plavras:

nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
# Criação de nova lista de palavras, removendo stopwords:

palavras_semstop = [p for p in palavras if p not in stops]
len(palavras_semstop)
```

```python
# Remoção da pontuação, grando uma lista sem stopwords e sem pontuação:

palavras_sem_pontuacao = [p for p in palavras_semstop if p not in string.punctuation]
len(palavras_sem_pontuacao)
```

```python
# Cáculo da frequência das palavras e visualização das mais comuns:

frequencia = nltk.FreqDist(palavras_sem_pontuacao)
frequencia
```

```python
# Mais comuns:

mais_comuns = frequencia.most_common(100)
mais_comuns
```
